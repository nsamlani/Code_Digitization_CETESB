```python
import fastwer
import jiwer
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import Image, display
import pandas
import pytesseract
from pytesseract import Output
import cv2
from statistics import mean


pytesseract.pytesseract.tesseract_cmd = 'C:\\Program Files\\Tesseract-OCR\\tesseract.exe'
```

```python
########### DATA 300 dpi ################
path_300 =  "Cropped 2012 Text/"
f_imgs = [f for f in os.listdir(path_300)]
str_nb = ['UTM_N','UTM_E']
str_lngTxt = ['Classification', 'Company' , 'Address']
str_shortTxt = ['Datum','Fuse']
# imgs_numbers = [f for f in os.listdir(path) if 'UTM' in f]
# imgs_lngTxt = [f for f in os.listdir(path) if any(s in f for s in str_lngTxt)]
# imgs_shortTxt = [f for f in os.listdir(path) if any(s in f for s in str_shortTxt)]

imgs_numbers = [f for f in os.listdir(path_300) if 'UTM' in f]
imgs_lngTxt = [f for f in os.listdir(path_300) if any(s in f for s in str_lngTxt)]
imgs_shortTxt = [f for f in os.listdir(path_300) if any(s in f for s in str_shortTxt)]
imgs_address = [f for f in os.listdir(path_300) if 'Address' in f]

len(imgs_lngTxt)
```

```python
########## Segnmentaion 
########## One line text = --psm 7 
# custom_config = r'--oem 2'


avgs_lngTxt = []
min_txt_lngTxt = []
empty_lngTxt_count = 0
scores = []
# config = '--psm 7'

# path_300 = "../Tesseract training/data2012_300dpi/"
# f_imgs = [f for f in os.listdir(path_300)]
for f in imgs_lngTxt:
    path_img =  path_300 + f

    img = cv2.imread(path_img,0)
#     img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    img = cv2.resize(img,(0,0),fx=1,fy=1)
#     denoised_img = cv2.fastNlMeansDenoising(img,None,9,13,11)
#     _, img_thresh = cv2.threshold(img, 10, 255, cv2.THRESH_BINARY)
#     img_smooth = cv2.medianBlur(img_thresh, 3)
#     img_float = img.astype(float)
#     gamma = 0.3
    
#     img_corrected = np.power(img_float, gamma)
#     img_corrected = np.uint8(img_corrected)
#     _, binary = cv2.threshold(img_corrected, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    ocr = pytesseract.image_to_data(img,lang='por',output_type = Output.DATAFRAME, pandas_config=None)
    confs = []
    for index, row in ocr.iterrows():
        if row['word_num'] != 0:
            confs.append((row['conf'],index))
            
    confs_dict_per_img = dict(confs)

    if len(confs_dict_per_img) != 0:
        avg_confs = mean(confs_dict_per_img.keys())
        avgs_lngTxt.append((f,avg_confs))

        min_conf_index = confs_dict_per_img[min(confs_dict_per_img.keys())]
        min_conf_txt = ocr.loc[ocr.index[min_conf_index],'text']
        min_txt_lngTxt.append((f,min_conf_txt))
              
    else:
        empty_lngTxt_count = empty_lngTxt_count + 1
        
avgs_lngTxt_dict=dict(avgs_lngTxt)
min_txt_lngTxt_dict=dict(min_txt_lngTxt)
print(empty_lngTxt_count)
```
